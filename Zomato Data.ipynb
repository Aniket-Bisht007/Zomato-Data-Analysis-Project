# Zomato Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

# Data Cleaning

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)
dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
dataframe.head()
```

# Chart1 (Type of Resturant)

```python
sns.countplot(x = dataframe['listed_in(type)'])
plt.xlabel("Type of resturant")
```

##### Conclusion = Majority of the resturant falls in dinning category

# Chart2 (Votes Per Resturant)

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c = "Green", marker = "o")
plt.xlabel("Type of Resturant",c = "Blue", size = 20)
plt.ylabel("Votes", c = "Blue", size = 20)
```

##### Conclusion = Dinning Resturants Recieved maximum votes

# Chart3 (Rating Distribution)

```python
plt.hist(dataframe['rate'],bins = 5)
plt.title("Rating Distribution")
plt.show()
```

##### Conclusion = Majority Resturants recived rating from 3.5 to 4

# Chart4 (Average Spending on order)

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x = couple_data)
```

##### Conclusion = Majority couples prefer resturants with an approx cost of 300/-

# Chart5 ( Mode to Receive max rating)

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

##### Conclusion = Offline Orders Recieve lower rating as compared to online orders.

# Chart6 (Orders Recieved by Resturants)

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order' , aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True, cmap = "YlGnBu", fmt = 'd')
plt.title("HeatMap")
plt.xlabel("Online Orders")
plt.ylabel("Listed in (Type)")
plt.show()
```

##### Conclusion = Dinning resturants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefers orders in person at resturants, but prefer online ordering at cafes.
